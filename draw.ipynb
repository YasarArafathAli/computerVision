```python
cv2.rectangle(img,(5,5),(495,95),(255,255,255),5)
cv2.rectangle(img,(200,200),(300,300),(255,255,0),5)
cv2.circle(img,(200,200),50,(255,0,255),5)
cv2.circle(img,(400,400),50,(255,0,255),-1)
cv2.line(img,(0,0),(500,500),(12,0,255),5)
```

```python
import cv2
import numpy as np


###########################
######## Function #########
###########################

def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y), 5, (255,255,255),-1)

cv2.namedWindow(winname='my_drawing')
cv2.setMouseCallback('my_drawing',draw_circle)

###########################
##Show image with opencv ##
###########################

img = np.zeros((800,800,3),np.int8)

while True:
    cv2.imshow('my_drawing',img)
    
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
